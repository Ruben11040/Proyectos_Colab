#### Descripción del problema

Tu objetivo es predecir qué tan probable es que las personas reciban sus vacunas contra la gripe H1N1 y la gripe estacional. Específicamente, estarás prediciendo dos probabilidades: una para h1n1_vaccine y otra para seasonal_vaccine.

Cada fila en el conjunto de datos representa a una persona que respondió a la Encuesta Nacional de la Gripe H1N1 de 2009.

**Variables a Obtener(Objetivos)**<br>
Para esta competición, hay dos variables objetivo:

- ***h1n1_vaccine*** - Si el encuestado recibió la vacuna contra la gripe H1N1.

- ***seasonal_vaccine*** - Si el encuestado recibió la vacuna contra la gripe estacional.

Ambas variables son binarias: 0 = No; 1 = Sí. Algunos encuestados no recibieron ninguna vacuna, otros solo recibieron una y algunos recibieron ambas. Esto se formula como un problema de múltiples etiquetas (y no multiclase).

**Las características en este conjunto de datos**<br>
Se proporciona un conjunto de datos con 36 columnas. La primera columna, respondent_id, es un identificador único y aleatorio. Las 35 características restantes se describen a continuación.

Para todas las variables binarias: 0 = No; 1 = Sí.

### Dendograma

Escalamos los valores de train para poder realizar el dendograma.

```python
scaler = StandardScaler()
train_scaled = scaler.fit_transform(dftrain.fillna(dftrain.mean()))
df_T = train_scaled.T
```

```python
# Para obtener el dendograma con el agrupamiento por características, se hace necesario obtener la transpuesta de la matriz del dataset

dist = sklearn.neighbors.DistanceMetric.get_metric('euclidean') 
D = dist.pairwise(df_T)
avD = np.average(D)
print("Distancia Media\t {:6.2f}".format(avD))

# 2. Constuimos el dendograma
plt.figure(figsize=(10, 6))
threshold_2 = 7
# Seleccionamos como distancia intercluster el vecino más alejado (complete)
clusters = cluster.hierarchy.linkage(D, method='complete')
cluster.hierarchy.dendrogram(clusters, color_threshold=threshold_2, labels=dftrain.columns.to_list(), leaf_rotation=90)
plt.show()
```

### AdaBoostClassifier para h1n1_vaccine

He elegido AdaBoostClassifier debido a que es el que mejor accuracy tiene, con un 0.84

```python
# Definir el espacio de búsqueda de los parámetros
param_dist = {
    "base_estimator__criterion" : ["gini", "entropy"],
    "base_estimator__splitter" : ["best", "random"],
    "base_estimator__max_depth" : range(1,20), # distribución uniforme entre 1 y 10
    "n_estimators": range(1,100),
    "learning_rate": [0.01, 0.1, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5, 7.0],
    "algorithm": ["SAMME", "SAMME.R"]
}
```

```python
# Crear el modelo con el mejor estimado base del modelo AdaBoostClassifier, que por lo que estuve leyendo este solo permite modelos que sean de árboles
base_estimator = DecisionTreeClassifier(random_state=42)

# Crear el estimador AdaBoost
ada = AdaBoostClassifier(base_estimator)
```

```python
# Realizar la búsqueda aleatoria de hiperparámetros

random_search = RandomizedSearchCV(ada, param_dist, n_iter=10, cv=5, scoring='accuracy')
random_search.fit(X_train_h1n1, y_train_h1n1)

# Imprimir los mejores parámetros encontrados y la puntuación
print("Mejores parámetros encontrados:")
print(random_search.best_params_)
print("Mejor puntuación:")
print(random_search.best_score_)
```

```python
# realizaremos una predición con el mejor modelo obtenido anteriormente
pred_h1n1 = random_search.best_estimator_.predict(X_test_h1n1)

#Evaluaremos el modelo usando el accuracy
score = accuracy_score(y_test_h1n1 ,pred_h1n1)

# imprimir precisión 
print(f'Los mejores parametros para predecir son: {random_search.best_estimator_}\n Accuracy --> {score}')
```

### Modelo de Salida multiple

Tras haber realizado un modelo para cada una de las variables a predecir, he decidido crear otro modelo que lo que hace es predecirte ambas variables a la vez.

```python
dftrain_y = dftrain[['h1n1_vaccine','seasonal_vaccine']]
dft = dftrain.drop(['h1n1_vaccine','seasonal_vaccine'],axis=1, inplace=False)
```

```python
s = scaler.fit_transform(dft)
```

```python
X_train, X_test, y_train, y_test = train_test_split(s, dftrain_y, test_size=0.3, random_state=42, shuffle = True)
```

He elegido el modelo de XGBClassifier ya que en los anteriores lazypredicts realizado, este modelo tiene un gran porcentaje de precisión en ambos valores. Fuera de este he utilizado un modelo de multietiqueta denominado MultiOutputClassifier y de modelo base he utilizado el mensionado al principio.

```python
# Definir los hiperparámetros a ajustar
param_dist = {
    'estimator__learning_rate': uniform(0.01, 0.2),  # Tasa de aprendizaje
    'estimator__n_estimators': range(50, 500),  # Número de estimadores (árboles)
    'estimator__max_depth': [3, 5, 7, 9],  # Profundidad máxima de los árboles
    'estimator__subsample': uniform(0.5, 0.5),  # Submuestreo de filas para construir cada árbol
    'estimator__colsample_bytree': uniform(0.5, 0.5),  # Submuestreo de columnas para construir cada árbol
    'estimator__reg_alpha': uniform(0, 1),  # Término de regularización L1
    'estimator__reg_lambda': uniform(0, 1),  # Término de regularización L2
}

# Crear el estimador base
base_estimator = XGBClassifier(random_state=42)

# Crear el modelo MultiOutputClassifier
multi_output_model = MultiOutputClassifier(base_estimator)

# Realizar la búsqueda aleatoria de hiperparámetros
random_search = RandomizedSearchCV(multi_output_model, param_dist, n_iter=10, cv=5, random_state=42)

# Entrenar el modelo con los datos de entrenamiento
random_search.fit(X_train, y_train)

# Obtener los mejores hiperparámetros y el mejor modelo
# random_search.best_params_
# random_search.best_estimator_
```

```python
# Evaluar el mejor modelo en los datos de prueba
y_pred = random_search.best_estimator_.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

# Calcular la matriz de confusión multietiqueta
mcm = multilabel_confusion_matrix(y_test, y_pred)

# Imprimir los resultados
print("Mejores hiperparámetros:", random_search.best_params_)
print("Accuracy:", accuracy)
```

```python
titulos = ["Vacuna H1N1", "Vacuna Estacional"]
# Visualizar la matriz de confusión multietiqueta
fig, axes = plt.subplots(2, 2, figsize=(10,10), sharex=True)
for i,(ax, matrix) in enumerate(zip(axes.flatten(), mcm)):
    sns.heatmap(matrix, annot=True, fmt="d", cmap="Blues", ax=ax, cbar=False)
    ax.set_title(titulos[i])
    ax.set_xlabel("Predicción")
    ax.set_ylabel("Verdad")
plt.tight_layout()
plt.show()
```
